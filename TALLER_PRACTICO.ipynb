# Operadores en Python

Rellenar: +, -, /, =, +=, -=, *, **, if, else, for

```python
# Operadores matemáticos sobre variables
a = 12
b = 10

print("a + b  =",a + b)
print("a * b  =",a * b)
print("a - b  =",a - b)
print("a ** b =",a ** b)  # a elevado a b (a^b hace algo completamente diferente)
print("a / b  =",a / b)   # Nota: en Python 2, / devuelve una división entera
print("a // b =",a // b)  # División entera (solo devuelve la parte entera de la división)
```

```python
a = 4
a -= 1    # De igual manera se puede usar -=, *=, /=
print(a)
```

```python
# Operadores Booleanos
a = (2 > 4)      # False 
b = (6 == 6)     # True

print("a        = ",a)
print("b        = ",b)
print("a or b   = ",a or b)    # False or True
print("a and b  = ",a and b)   # False and True
```

```python
# Se puede dejar en blanco el inicio o el fin
print('x      :', x)
print('x[:3]  :',x[:3])      # Índices 0, 1, 2 (sin incluir el índice 3)
print('x[2:]  :',x[2:])      # Índices 2, 3, ... (hasta el final) 
print('x[:]   :',x[:])       # Es la lista completa
print('x[:-1] :',x[:-1])     # Índices 0, 1, ..., -2 (no incluye el índice -1 que es el último elemento)
```

# Face detection

```python
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml
```

```python
import cv2
import numpy as np

# Load the cascade
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Read the input image
from skimage import io

img = io.imread('https://static01.nyt.com/images/2020/11/19/us/artificial-intelligence-fake-people-faces-promo-1605818328743/artificial-intelligence-fake-people-faces-promo-1605818328743-superJumbo-v2.jpg')
print(img.shape)

fig = plt.figure(figsize=(20,10))
plt.imshow(img)
plt.show()
```

```python
# Convert into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the output
fig = plt.figure(figsize=(20,10))
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
# Detect faces
faces = face_cascade.detectMultiScale(gray, 1.1, 4)
faces
```

```python

# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Display the output
fig = plt.figure(figsize=(20,10))
plt.imshow(img)
plt.show()
```
